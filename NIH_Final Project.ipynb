```python
#2. Set column names to lowercase

# Get the current user's home directory
home_directory = os.path.expanduser("~")

# Define the absolute path to the input directory
absolute_directory_path = os.path.join(home_directory, "Documents", "nih")

# Define the absolute path to the output folder
absolute_output_folder = os.path.join(home_directory, "Documents", "nih_cleaned_dfs")

# Function to fix the case of column names to lowercase and save to individual CSV files
def fix_case_and_save(directory_path, output_folder):
    print("\nFixing Case:")

    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Iterate through each file in the directory
    for file_name in os.listdir(directory_path):
        if file_name.endswith(".csv"):
            # Construct the full file path
            file_path = os.path.join(directory_path, file_name)

            try:
                # Read the CSV file into a DataFrame with 'ISO-8859-1' encoding
                df = pd.read_csv(file_path, encoding='ISO-8859-1', on_bad_lines='skip')

                # Fix the case: Change column names to lowercase
                df.columns = [column.lower() for column in df.columns]

                # Display the fixed DataFrame
                print(df)

                # Save the fixed DataFrame to a new CSV file in the output folder
                save_path = os.path.join(output_folder, f"{file_name.replace('.csv', '_cleaned.csv')}")
                df.to_csv(save_path, index=False)

                print(f"Saved fixed DataFrame to: {save_path}\n")

            except UnicodeDecodeError as e:
                print(f"Error reading file {file_name}: {e}")

# Call the fix_case_and_save function
fix_case_and_save(absolute_directory_path, absolute_output_folder)

# Print confirmation message
print("Fixing case completed successfully.")


```

```python
#9.Check for NAs, etc.

# Get the current user's home directory
home_directory = os.path.expanduser("~")

# Define the relative path to the cleaned CSV files
relative_cleaned_directory_path = "Documents/nih_cleaned_dfs"

# Combine the home directory with the relative cleaned directory path to get the full path
cleaned_directory_path = os.path.join(home_directory, relative_cleaned_directory_path)

# Function to filter columns by missing percentage
def filter_columns_by_missing_percentage(file_path, min_threshold=0.00, max_threshold=100):
    # Read the DataFrame from the CSV file
    df = pd.read_csv(file_path)

    columns_info = []

    # Loop through each column in the DataFrame
    for column in df.columns:
        if column != 'fy':
            missing_percentage = df[column].isna().mean() * 100
            if min_threshold <= missing_percentage <= max_threshold:
                if missing_percentage > 0.00:  # Check if missing percentage is greater than 0.00
                    columns_info.append({'DataFrame': os.path.splitext(os.path.basename(file_path))[0], 'Column': column, 'MissingPercentage': missing_percentage})

    # Print columns with missing data greater than 0.00%
    print(f"\nColumns with missing data greater than 0.00% in {os.path.basename(file_path)}:")
    for item in columns_info:
        print(f"Column: {item['Column']}, DataFrame: {item['DataFrame']}, Missing Percentage: {item['MissingPercentage']:.2f}%")

# Apply the function to each CSV file in the directory
for file_name in os.listdir(cleaned_directory_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(cleaned_directory_path, file_name)
        filter_columns_by_missing_percentage(file_path, min_threshold=0.00, max_threshold=100)
```
